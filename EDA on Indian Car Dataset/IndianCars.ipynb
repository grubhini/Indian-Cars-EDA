```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cars_ds_final.csv')
df.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),cbar=False,cmap="viridis")
```

```python
for c in df.columns:
  if(df[c].isnull().sum()>625):
    df.drop(columns=[c],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
for c in df.columns:
  unique=df[c].nunique()
  print(f'{c}:{unique}')
  if(unique==1):
    df.drop(columns=[c],inplace=True,axis=1)
  if(unique==df.shape[0]):
    df.drop(columns=[c],inplace=True,axis=1)
```

```python
df.head()
```

```python
for c in df.columns:
  print(f'{c}:{df[c].dtype}:{df[c].isnull().sum()}')
```

```python
new_df=df.copy()
```

```python
df=new_df.copy()
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df['Ex-Showroom_Price']=df['Ex-Showroom_Price'].str.replace('Rs. ','')
df['Ex-Showroom_Price']=df['Ex-Showroom_Price'].str.replace(',','').astype(int)
```

```python
df['Ex-Showroom_Price']
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
row_index=df.index[df['Make'].isnull()].tolist()
print(row_index)
```

```python
for i in row_index:
  df['Make'].iloc[i]=df['Model'].iloc[i].split()[0]
```

```python
df.iloc[252]
```

```python
df['Make'].replace('Go+','Datsun',inplace=True)
```

```python
df.iloc[875]
```

###Converting the fields into appropriate data types

```python
new_df.drop_duplicates(keep='first',inplace=True)
```

```python
for c in ['Height','Length','Width']:
  df[c]=new_df[c]
```

```python
for c in ['Height','Length','Width']:
  df[c]=df[c].str.replace(' mm','')
  df[c]=pd.to_numeric(df[c],errors='raise')
```

###Plot scatter plots to explore the relationship between two numeric variables (e.g., sales vs. price).
